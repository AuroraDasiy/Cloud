### 练习
1. 读入stock0.txt中的数据，每列数据之间以制表符分隔。其中，一行一只股票，第一列是前一天的收盘价，第二列是当天的收盘价
2. 计算每只股票的涨跌值与涨跌幅，涨跌幅=涨跌值/前一天的收盘价
3. 合并数据，组成(6000,4)数组，每列分别是 前一天的收盘价、当天的收盘价、涨跌值、涨跌幅
4. 设置以定点法print数据，只精确到2位小数。查询前10只股票的情况
5. 查询涨幅高于10%的股票
6. 查询前20只股票的当天收盘价与涨跌幅
7. 已知A购买了前20只股票，各股票的数量按顺序为：
    100,200,300,150,500,600,700,550,900,1000,100,100,100,200,300,800,550,150,900,1000
    
    请计算A当天的所购股票的总市值、总收益
    
    (正确结果是：总市值 182955.5 总收益 480.50000000000284）
8. 求购买股票的数量总和、最小数量

    （正确结果是：股票总数量 9200 最小数量 100）
9. 将股票按涨跌幅降序排序，排序结果如下：

~~~
 [[ 29.5   33.99   4.49   0.15]
 [139.82 155.36  15.54   0.11]
 [ 85.5   95.     9.5    0.11]
 ...
 [ 56.72  47.27  -9.45  -0.17]
 [ 50.83  42.36  -8.47  -0.17]
 [ 46.48  38.73  -7.75  -0.17]]
 ~~~

```python
import numpy as np

# 假设 stock0.txt 文件的路径是 'path_to_stock0.txt'
# 请将 'path_to_stock0.txt' 替换为实际的文件路径
file_path = 'stock0.txt'

# 读取数据，假设每列之间以制表符分隔
data = np.loadtxt(file_path, delimiter='\t',encoding='utf-16')

# 计算涨跌值和涨跌幅
price_change = data[:, 1] - data[:, 0]
percentage_change = price_change / data[:, 0]

# 合并数据
combined_data = np.column_stack((data, price_change, percentage_change))

# 设置定点法打印数据，精确到2位小数
np.set_printoptions(precision=2, suppress=True)

# 查询前10只股票的情况
print(combined_data[:10])

# 查询涨幅高于10%的股票
high_growth_stocks = combined_data[combined_data[:, 3] > 0.10]

# 查询前20只股票的当天收盘价与涨跌幅
print(combined_data[:20, [1, 3]])

# 计算A的总市值和总收益
quantities = np.array([100, 200, 300, 150, 500, 600, 700, 550, 900, 1000, 100, 100, 100, 200, 300, 800, 550, 150, 900, 1000])
total_value = np.dot(combined_data[:20, 1], quantities)
total_profit = np.dot(price_change[:20], quantities)

# 输出A的总市值和总收益
print(f'总市值: {total_value:.2f}')
print(f'总收益: {total_profit:.2f}')

# 股票总数量和最小数量
total_stocks = quantities.sum()
min_quantity = quantities.min()

# 输出股票总数量和最小数量
print(f'股票总数量: {total_stocks}')
print(f'最小数量: {min_quantity}')

# 按涨跌幅降序排序
sorted_stocks = combined_data[combined_data[:, 3].argsort()[::-1]]

# 打印排序后的股票信息
print(sorted_stocks)
```
